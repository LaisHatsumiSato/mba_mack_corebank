```python
%sql
USE corebank_catalog.gold_corebank
```

```python
%sql
CREATE OR REPLACE VIEW vw_dash_fraude_pix AS
SELECT
  dia_transacao,
  COUNT(*)                                       AS qtd_transacoes,
  SUM(valor_transacao)                           AS valor_total,
  AVG(valor_transacao)                           AS ticket_medio,
  SUM(CASE WHEN risco_fraude = 'ALTO_RISCO' THEN 1 ELSE 0 END) AS qtd_suspeitas,
  SUM(CASE WHEN risco_fraude = 'ALTO_RISCO' THEN valor_transacao ELSE 0 END) AS valor_suspeito
FROM corebank_catalog.gold_corebank.gold_fraude_pix
GROUP BY dia_transacao;
```
