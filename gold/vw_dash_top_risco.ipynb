```python
%sql
USE corebank_catalog.gold_corebank
```

```python
%sql
CREATE OR REPLACE VIEW vw_dash_top_risco AS
SELECT
  id_cliente,
  COUNT(*)                                         AS qtd_trans_suspeitas,
  SUM(valor_transacao)                             AS valor_suspeito,
  MAX(dt_hr_transacao)                             AS ultima_suspeita
FROM corebank_catalog.gold_corebank.gold_fraude_pix
WHERE risco_fraude = 'ALTO_RISCO'
GROUP BY id_cliente
ORDER BY valor_suspeito DESC;
```
