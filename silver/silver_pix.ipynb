```python
%sql
CREATE OR REPLACE TABLE corebank_catalog.silver_corebank.silver_transacao_pix AS
SELECT
  id_transacao,
  id_cliente,
  id_banco,
  id_banco_destino,
  id_cliente_destino,
  CAST(valor_transacao AS DOUBLE)                     AS valor_transacao,
  TO_TIMESTAMP(dt_hr_transacao)                       AS dt_hr_transacao,
  DATE(TO_TIMESTAMP(dt_hr_transacao))                 AS dia_transacao,
  HOUR(TO_TIMESTAMP(dt_hr_transacao))                 AS hora_transacao
FROM corebank_catalog.bronze_corebank.transacao_pix
WHERE id_transacao IS NOT NULL;
```
