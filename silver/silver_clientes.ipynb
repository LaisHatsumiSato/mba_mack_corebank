```python
%sql
CREATE OR REPLACE TABLE corebank_catalog.silver_corebank.silver_clientes_anonimizados AS
SELECT
  id_cliente,
  id_banco,
  estado,
  cidade,
  SHA2(cpf_cnpj, 256) AS cpf_hash,
  REGEXP_REPLACE(telefone, '(\\d)(\\d{4,5})(\\d)', '$1****$3') AS telefone_mask,
  REGEXP_REPLACE(email, '(.+)@(.+)', '****@$2') AS email_mask,
  FLOOR(DATEDIFF(current_date(), TO_DATE(data_nascimento, 'yyyy-MM-dd')) / 365.25) AS idade,
  chave_pix
FROM corebank_catalog.bronze_corebank.clientes
WHERE id_cliente IS NOT NULL;
```
