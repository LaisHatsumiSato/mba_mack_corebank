```python
%sql
USE corebank_catalog.gold_corebank;
```

```python
%sql
SELECT
  dia_transacao,
  qtd_transacoes,
  qtd_suspeitas,
  valor_suspeito
FROM vw_dash_fraude_pix
WHERE dia_transacao >= date_add(current_date(), -1)
ORDER BY dia_transacao
LIMIT 30;
```

```python
%sql
SELECT
  dia_transacao,
  qtd_transacoes,
  qtd_suspeitas,
  valor_suspeito
FROM vw_dash_fraude_pix
WHERE dia_transacao >= date_add(current_date(), -1)
ORDER BY dia_transacao
LIMIT 30;
```

```python
%sql
SELECT
  SUM(qtd_transacoes)  AS total_transacoes,
  SUM(qtd_suspeitas)   AS total_suspeitas,
  SUM(valor_suspeito)  AS total_valor_suspeito
FROM vw_dash_fraude_pix
WHERE dia_transacao = '2025-11-30';
```

```python
%sql
SELECT
  id_cliente,
  qtd_trans_suspeitas,
  valor_suspeito,
  ultima_suspeita
FROM vw_dash_top_risco
WHERE DATE(ultima_suspeita) = '2025-11-30'
ORDER BY valor_suspeito DESC
LIMIT 10;
```

```python
%sql
SELECT
  dia_transacao,
  qtd_transacoes,
  qtd_suspeitas
FROM vw_dash_fraude_pix
WHERE dia_transacao = '2025-11-30'
ORDER BY dia_transacao;
```

```python
%sql
SELECT
  hora_transacao,
  COUNT(*) AS qtd_transacoes,
  SUM(CASE WHEN risco_fraude = 'ALTO_RISCO' THEN 1 ELSE 0 END) AS qtd_suspeitas
FROM corebank_catalog.gold_corebank.gold_fraude_pix
WHERE dia_transacao = '2025-11-30'
GROUP BY hora_transacao
ORDER BY hora_transacao;
```
