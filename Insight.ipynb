```python
%sql
USE corebank_catalog.gold_corebank;
```

```python
%sql
SELECT
  dia_transacao,
  qtd_transacoes,
  qtd_suspeitas,
  valor_suspeito
FROM vw_dash_fraude_pix
WHERE dia_transacao >= date_add(current_date(), -2)
ORDER BY dia_transacao
LIMIT 30;
```

```python
%sql
SELECT
  dia_transacao,
  valor_suspeito
FROM vw_dash_fraude_pix
WHERE dia_transacao >= date_add(current_date(), -9)
ORDER BY dia_transacao;
```

```python
%sql
SELECT
  id_cliente,
  qtd_trans_suspeitas,
  valor_suspeito
FROM vw_dash_top_risco
ORDER BY valor_suspeito DESC
LIMIT 10;
```

```python
%sql
SELECT *
FROM vw_dash_top_risco
ORDER BY valor_suspeito DESC;
```

```python
%sql
SELECT 
  COUNT(*) AS total,
  SUM(CASE WHEN risco_fraude = 'ALTO_RISCO' THEN 1 ELSE 0 END) AS suspeitas,
  ROUND(100.0 * SUM(CASE WHEN risco_fraude = 'ALTO_RISCO' THEN 1 ELSE 0 END) / COUNT(*), 2) AS pct_suspeitas
FROM gold_fraude_pix;
```
